## __Image Contours__

Image contours are curves joining all the continuous points (along the boundary) that have the same color or intensity. Contours are very useful for:

Shape analysis

Object detection and recognition

Image segmentation

## __CONTOUR MOMENTS AND CONVEXHULL__

Contour Moments
Moments are quantitative measures of the shape and spatial distribution of contours. They help in calculating features like centroid, area, orientation, etc.


Convex Hull
A convex hull is the smallest convex polygon that can contain all the points of a contour.

```python
import cv2
import numpy as np

img = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\shape.png')
img = cv2.resize(img,(500,500))

gry = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

r,thre = cv2.threshold(gry,220,255,cv2.THRESH_BINARY)

contour,hax = cv2.findContours(thre,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)



# isse hm center , area 

ar = []

for i in contour:
    m = cv2.moments(i)
    x = int(m['m10']/m['m00'])
    y = int(m['m01']/m['m00'])
    cv2.drawContours(img,contour,-1,(0,255,0),2)
    cv2.circle(img,(x,y),4,(0,0,0),-1)

    a = cv2.contourArea(i)
    ar.append(a)

    # convex hull 

    ep = 0.1*cv2.arcLength(i,True)
    d = cv2.approxPolyDP(i,ep,True)
    hull = cv2.convexHull(d)
    x,y,w,h = cv2.boundingRect(hull)
    cv2.rectangle(img,(x,y),(x+w,y+h),(0,255,0),3)


print(ar)



# cv2.imshow('imagethre',thre)
cv2.imshow('imagethre',thre)
cv2.imshow('image',img)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

def nothing(x):
    pass

cv2.namedWindow('wscube')
cv2.createTrackbar('Threshold','wscube',0,255,nothing)

cv2.createTrackbar('lb','wscube',0,255,nothing)
cv2.createTrackbar('lg','wscube',0,255,nothing)
cv2.createTrackbar('lr','wscube',0,255,nothing)
cv2.createTrackbar('hb','wscube',0,255,nothing)
cv2.createTrackbar('hg','wscube',0,255,nothing)
cv2.createTrackbar('hr','wscube',0,255,nothing)


cap = cv2.VideoCapture(0)

while cap.isOpened():
    ret,frame = cap.read()
    frame = cv2.flip(frame,1)
    frame = cv2.resize(frame,(200,200))

    if ret == True:

        
        LB = cv2.getTrackbarPos('lb','wscube')
        LG  = cv2.getTrackbarPos('lg','wscube')
        LR = cv2.getTrackbarPos('lr','wscube')
        hB = cv2.getTrackbarPos('hb','wscube')
        hG  = cv2.getTrackbarPos('hg','wscube')
        hR = cv2.getTrackbarPos('hr','wscube')
        hsv = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)

        lower = np.array([LB,LG,LR])
        upper = np.array([hB,hG,hR])
        
        th = cv2.getTrackbarPos('Threshold','wscube')
        m = cv2.inRange(hsv,lower,upper)
        res =cv2.bitwise_and(frame,frame,mask=m)
        fr = cv2.bitwise_not(res)

        _,thi = cv2.threshold(m,th,255,cv2.THRESH_BINARY)
        cnt,hr = cv2.findContours(thi,cv2.RETR_TREE,cv2.CHAIN_APPROX_SIMPLE)

        cv2.drawContours(frame,cnt,-1,(0,255,0),2)


        cv2.imshow('threr',thi)
        cv2.imshow('ress',res)
        cv2.imshow('mask',m)
        cv2.imshow('lie',hsv)
        cv2.imshow('live',frame)
        if cv2.waitKey(25) == ord('q'):
            break


    else:
        break


cap.release()
cv2.destroyAllWindows()
```

## Hough transformation lines

it is used for line detection ,if line is broken than  it is used 

y = mx + c 


r cos Î± + r sin Î± = p

```python
import cv2
import numpy as np

image = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\istockphoto-1291290878-612x612.jpg')

gry = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

edge = cv2.Canny(gry,20,250)

lines = cv2.HoughLines(edge,1,np.pi/180,200)
print(lines)

for r,the in lines[0]:
    a = np.cos(the)
    b = np.sin(the)

    x0 = a*r
    y0 = b*r

    x1 = int(x0+1000*(-b))
    y1 = int(y0+1000*(a))
    x2 = int(x0-1000*(-b))
    y2 = int(y0-1000*(a))

    cv2.line(image,(x1,y1),(x2,y2),(0,255,0),2)


cv2.imshow('wscode',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\Man and woman jogging together on the street  1080p HD video  Copyright Free  Stock Videos - Stock Vlog (1080p, h264) (1).mp4')

ret,frame = cap.read()
x,y,w,h =400,99,400,885   # isko phle paint me jakr or fir apne hisab se adjust kr k dekhna pdega 

t = (x,y,w,h)
roi = frame[y:y+h,x:x+w]
hsv_roi = cv2.cvtColor(roi,cv2.COLOR_BGR2HSV)
mask = cv2.inRange(hsv_roi,np.array((0.,60.,32.)),np.array((180.,255.,255.)))
roi_hist = cv2.calcHist([hsv_roi],[0],mask,[180],[0,180])
cv2.normalize(roi_hist,roi_hist,0,255,cv2.NORM_MINMAX)
cv2.imshow('test',roi)

# termination point -- tr
tr = (cv2.TermCriteria_EPS | cv2.TERM_CRITERIA_COUNT,10,1)



while cap.isOpened():
    ret , frame = cap.read()
    
    if ret == True:
        hsv_frame = cv2.cvtColor(frame,cv2.COLOR_BGR2HSV)
        # Search backimage
        d = cv2.calcBackProject([hsv_frame],[0],roi_hist,[0,180],1)
        r ,tp = cv2.meanShift(d,t,tr)
        x,y,w,h = tp
        final = cv2.rectangle(frame,(x,y),(x+w,y+h),(0,0,255),4)
        
        # cv2.imshow('Thisisframe',frame)
        cv2.imshow('Thisisframe',final)

        if cv2.waitKey(25) == ord('p'):
            break
    else:
        break


cap.release()
cv2.destroyAllWindows()
```

## Corner Detection using Detection Method

## Harris corner detection

```python
# __Harris corner Detection__
#  jnha 2 edges milte h vo wale corner ko detect krta h ( junction ko detect krta haiiii)
import cv2
import numpy as np

# image = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\shape.png')
image = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\1734083369576-i.webp')
image = cv2.resize(image,(500,500))
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)


# isme aapko integer ko float me convert krna pdega

gray = np.float32(gray)

resut = cv2.cornerHarris(gray,4,3,0.04)

resut = cv2.dilate(resut,None)
print(resut)

image[resut>0.01*resut.max()] = [0,0,255]
 
# print(gray)
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np


image = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\shape.png')
image = cv2.resize(image,(500,500))
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

corner = cv2.goodFeaturesToTrack(gray,25,0.01,20)
# print(corner)

# convert this points in interger bc they are in float 

corner = np.int64(corner)

for i in corner:
    # revel convert  into 2D to 1D
    x,y = i.ravel()
    cv2.circle(image,(x,y),3,(0,0,255),-1)

cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

```python
import cv2
import numpy as np

image = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\1734083369576-i.webp')
gray = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)

#  this is the file for detection for face
face = cv2.CascadeClassifier(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\.venv\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml')

detect = face.detectMultiScale(gray,1.2,4)  # 1.3 is the distance from the face
# detection give the points of faces 
# print(detect)
count = 0
for (x,y,w,h) in detect:
    count +=1
    cv2.rectangle(image,(x,y),(x+w,y+h),(0,0,255),3)

print('This is the count of faces in image = ',count)
cv2.imshow('image',image)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

## __Reverse the video__

```python
import cv2
import numpy as np

cap = cv2.VideoCapture(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\Man and woman jogging together on the street  1080p HD video  Copyright Free  Stock Videos - Stock Vlog (1080p, h264) (1).mp4')

# file name ko list me store krege

listt = []
c = 0

while cap.isOpened():
    ret , frame = cap.read()
    # frame = cv2.resize(frame,(600,400))

    if ret == True:
        c+=1
        file_name = r"C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\images\demo"+str(c)+".jpg"
        listt.append(file_name)

        # video ko reverse krne se phle video ki sari image dusre folder me store krni pdegi 
        cv2.imwrite(file_name,frame)

        cv2.imshow('wscube',frame)
        if cv2.waitKey(25) == ord('p'):
            break
    
    else:
        break

listt.reverse()
for i in listt:
    image = cv2.imread(i)
    # image = cv2.resize(image,(600,400))
    cv2.imshow('new',image)
    if cv2.waitKey(25) == ord('p'):
        break

cap.release()
cv2.destroyAllWindows()
```

```python

import cv2
import numpy as np


image  = cv2.imread(r'C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\1734083369576-i.webp')
gray = cv2.cvtColor(image,cv2.COLOR_BGRA2GRAY)
harcaste_eye = cv2.CascadeClassifier(r"C:\Users\Lenovo\Downloads\haarcascade_eye.xml")
harcaste_face = cv2.CascadeClassifier(r"C:\Users\Lenovo\Krishnaraj singh\Code\Opencv\.venv\Lib\site-packages\cv2\data\haarcascade_frontalface_default.xml")

face = harcaste_face.detectMultiScale(gray,1.3,2)

for x,y,w,h in face:
    roi_gray = gray[y:y+h,x:x+w]
    roi_image = image[y:y+h,x:x+w]

    s = harcaste_eye.detectMultiScale(roi_image,1.3,3)

    for xs,ys,ws,hs in s:
        cv2.rectangle(roi_image,(xs,ys),(xs+ws,ys+hs),(0,255,0),2)


cv2.imshow('new',roi_gray)
cv2.imshow('nmnm',roi_image)
cv2.imshow('nmnmdf',image)

cv2.waitKey(0)
cv2.destroyAllWindows()
```
